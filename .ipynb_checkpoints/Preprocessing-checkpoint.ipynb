```python
url = r'C:\Users\nitinkohli\MIAS\all-mias'

df = pd.read_csv(r'C:\Users\nitinkohli\MIAS\Info.txt', sep=' ')

img_name = []

for i in range(df.shape[0]):
        img_name.append(df.REFNUM[i]+ '.pgm')
img_name = np.array(img_name)
        
img_add = []
for i in range(df.shape[0]):
    img_add.append(url + '\\' + df.REFNUM[i]+ '.pgm') 
img_add = np.array(img_add)
```

```python
outdir = r'C:\Users\nitinkohli\MIAS\labels_removed'
import os
import random
x = 5
y = 4
p = 300
for i in range(0 + p, 5 + p):
    fig = plt.figure(figsize = (25, 15))
    
    rand = random.randint(0, 330)
#     img = cv2.imread(img_add[i], 0)
    img = cv2.imread(img_add[rand], 0)
    
    nrows, ncols = img.shape
#     img = img[1:nrows, 0:ncols]
    img = img[10:nrows-10, 15:ncols-15]
    
    
    ax = plt.subplot(x, y, (i - p) * 4 + 1)
    plt.title("orig")
    plt.imshow(img, cmap="gray")
    
    ret, thresh1 = cv2.threshold(img, 20, 255, cv2.THRESH_BINARY)
    
    ax = plt.subplot(x, y, (i - p) * 4 + 2)
    plt.title("thresh")
    plt.imshow(thresh1, cmap="gray")
    
    im, contours, hierarchy = cv2.findContours(thresh1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE )
    maxContour = 0
    for contour in contours:
        contourSize = cv2.contourArea(contour)
        if contourSize > maxContour:
            maxContour = contourSize
            maxContourData = contour
    mask = np.zeros_like(thresh1)
    cv2.fillPoly(mask,[maxContourData],1)
    mask1 = cv2.GaussianBlur(mask, (49,49), 0)
    mask = cv2.medianBlur(mask1, 49)
    ax = plt.subplot(x, y, (i - p) * 4 + 3)
    plt.title("mask")
    plt.imshow(mask, cmap="gray")
    
    img = cv2.bitwise_and(img, img, mask = mask)
    
    ax = plt.subplot(x, y, (i - p) * 4 + 4)
    plt.title("final")
    plt.imshow(img, cmap="gray")
    
#     outfile = img_name[i]
#     cv2.imwrite(os.path.join(outdir, outfile), img)
```
